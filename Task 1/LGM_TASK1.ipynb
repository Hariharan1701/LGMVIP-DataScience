# **TRAINING & TESTING**

```python
# Spliting the data for testing 
Label_Encode = LabelEncoder()
y= df['species']
y= Label_Encode.fit_transform(y)
y
```

```python
x=df.iloc[:,:-1].values
x
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.33, random_state=5)
```

## **LOGISTIC REGRESSION**

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(random_state=3)
model.fit(xtrain,ytrain)
```

```python
y_pred = model.predict(xtest)
```

```python
# predicted values
y_pred
```

```python
# actual values
ytest
```

```python
mymodel=pd.DataFrame(y_pred,ytest).reset_index()
mymodel.columns = ['tested','predicted']
mymodel.head()
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(ytest,y_pred)
print("CONFUSION MATRIX: ")
print (cm)

print()

print("ACCURACY SCORE OF LOGISTIC REGRESSION MODEL:",accuracy_score(ytest,y_pred))
```

## **KMEANS CLUSTERING**

```python
from sklearn.cluster import KMeans
```

**ELBOW METHOD (TO CHOOSE OPTIMAL NO. OF CLUSTERS)**

```python
li=[]
cluster_range=range(1,10)
for i in cluster_range:
  km=KMeans(n_clusters=i)
  km.fit(x)
  inertia=km.inertia_
  li.append(inertia)
print(li)

plt.plot(cluster_range,li)
plt.show()
```

```python
km= KMeans(n_clusters=3,random_state=2)
km.fit(x)
```

```python
km.cluster_centers_
```

```python
labels= km.labels_
labels
```

```python
ypred_k=km.predict(xtest)
ypred_k
```

```python
ytest
```

```python
print("Accuracy: ", accuracy_score(ytest,ypred_k))
```

```python
color=np.array(['Blue','orange','green'])
plt.scatter(df['sepal_length'],df['sepal_width'],20, c=color[labels])
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.show()
```

# **DECISION TREE**

```python
from sklearn import tree
d_tree=tree.DecisionTreeClassifier()
d_tree.fit(xtrain,ytrain)
```

```python
ypred_dtree = d_tree.predict(xtest)
```

```python
print("Accuracy: ", accuracy_score(ytest,ypred_dtree))
print()

print("Confusion Matrix: ")
conf_matrix=confusion_matrix(ytest,ypred_dtree)
conf_matrix
```
