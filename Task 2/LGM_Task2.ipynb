## **IMPORT THE LIBRARIES**





```python
#import the library
from google.colab.patches import cv2_imshow
import cv2
```

## **READ THE IMAGE**

```python
# Load and read the image
image=cv2.imread('/content/image.jpg')
cv2_imshow(image)
```

## **CONVERT THE IMAGE INTO GRAYSCALE**

```python
# Convert the original(RGB) image into grayscale
gray_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

## **INVERT THE GRAYSCALE IMAGE**


```python
# Invert the grayscale image to get a negative
invgray_image = 255 - gray_image
cv2_imshow(invgray_image)
```

## **BLUR  THE IMAGE**

```python
# Apply a Gaussian blur to the negative
blur_image = cv2.GaussianBlur(invgray_image, (21, 21), 0)
cv2_imshow(blur_image)
```

## **MIXING THE GRAYSCALE IMAGE & INVERTED BLUR IMAGE**

```python
# Blend the grayscale image  with the blurred negative using a color dodge
def dodge(gray_image, blur_image):
    invblur_img = 255 - blur_image
    return cv2.divide(gray_image, invblur_img, scale = 256)
```

```python
pencil_sketch = dodge(gray_image, blur_image)
cv2_imshow(pencil_sketch)
```
